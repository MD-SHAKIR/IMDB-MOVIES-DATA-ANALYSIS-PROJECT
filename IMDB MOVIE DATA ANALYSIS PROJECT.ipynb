## Subtask 3.1
#### 1. Create  a data_by_genre dataframe which should have columns genre1,genre2,genre3 and all the columns related to Cvotes and Votes
#### 2. Create a column by count and initialiser it to 1

```python
#Created a dataframe named df_by_genre
col_list=[]
for i in movies.columns:
    if i.startswith("CV")|i.startswith("V")|i.startswith("ge"):
        col_list.append(i)
```

```python
movies.loc[:,col_list]
```

```python
df_by_genre= movies.loc[:,col_list]
df_by_genre.head()
```

```python
#Created a column in dataframe by cnt(count) and initialise to 1 
df_by_genre["cnt"]=1
df_by_genre
```

```python
#Here, we use GROUP BY function,and with groupby function we are aggregate function also
#Then we store this in a new dataframe by name ------ df_by_g1

df_by_g1=df_by_genre.groupby(by="genre_1").sum()
```

```python
df_by_g1
```

```python
#Same with Genre_2
df_by_g2=df_by_genre.groupby(by="genre_2").sum()
df_by_g2
```

```python
#Same with Genre_3
df_by_g3=df_by_genre.groupby(by="genre_3").sum()
df_by_g3
```

```python
#Then add the grouped dataframe(genre_1 & genre_2) and store it in a new dataframe

df1_df2=df_by_g1.add(df_by_g2,fill_value=0)
```

```python
df1_df2
```

```python
#and then add dataframes (df1_df2 & genre_3) and store in a df_final dataframe
df_final=df1_df2.add(df_by_g3,fill_value=0)
```

```python
df_final
```

```python
#Find The Top 10 genre where count is more or equal to 10
top_10_genre=df_final[df_final["cnt"] >=10]
```

```python
top_10_genre
```

```python
#Taking The MEAN for every column by dividing with "cnt "column.
cnt_column=top_10_genre['cnt']
top_10_genre=top_10_genre.div(cnt_column,axis='index')
```

```python
top_10_genre
```

```python
#Rounding off the columns of values to two decimal,then store in same dataframe
top_10_genre =np.round(top_10_genre,2)
top_10_genre
```

```python
#Converting "CVotes" columns to integer type first of all ,we a code for loop
cvotes_list=[]
for i in top_10_genre.columns:
    if i.startswith("CVotes"):
        cvotes_list.append(i)
```

```python
cvotes_list
```

```python
#Converting column with CVotes from float to int
top_10_genre[cvotes_list]=top_10_genre[cvotes_list].astype('int')
```

```python
top_10_genre
```

##  Subtask 3.4: " US " vs " Non-US " Cross Analysis

The dataset contains both the US and non-US movies. Let's analyse how both the US and the non-US voters have responded to the US and the non-US movies.

1. Create a column `IFUS` in the dataframe `movies`. The column `IFUS` should contain the value "USA" if the `Country` of the movie is "USA". For all other countries other than the USA, `IFUS` should contain the value `non-USA`.


2. Now make a boxplot that shows how the number of votes from the US people i.e. `CVotesUS` is varying for the US and non-US movies. Make use of the column `IFUS` to make this plot. Similarly, make another subplot that shows how non US voters have voted for the US and non-US movies by plotting `CVotesnUS` for both the US and non-US movies. Write any of your two inferences/observations from these plots.


3. Again do a similar analysis but with the ratings. Make a boxplot that shows how the ratings from the US people i.e. `VotesUS` is varying for the US and non-US movies. Similarly, make another subplot that shows how `VotesnUS` is varying for the US and non-US movies. Write any of your two inferences/observations from these plots.

Note : Use `movies` dataframe for this subtask. Make use of this documention to format your boxplot - https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
#Creating IFUS columns
movies['IFUS']=movies['Country'].apply(lambda x: 'USA' if x=='USA' else 'Non-USA')
movies
```

```python
# Box plot - 1: CVotesUS(y) vs IFUS(x)

fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(15,8))
fig.subplots_adjust(wspace=0.01)
sns.boxplot(x='IFUS',y='CVotesUS',data=movies,ax=ax1)
plt.title('NON-USA Votes for USA & NON-USA movies')
sns.boxplot(x='IFUS',y='CVotesnUS',data=movies,ax=ax2)
plt.tight_layout()
plt.show()
```

**`Inferences:`** Here are the two inferences/observations below:
- Inference 1:USA movies have got more votes than non USA movies from both category of people i.e from USA and not from USA.
- Inference 2:USA people have voted more for USA movies comparing it with non-USA movies.
- Inference 3: non-USA people have voted more for USA movies comparing it with non-USA movies.

```python
# Box plot - 2: VotesUS(y) vs IFUS(x)

fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(15,8))
fig.subplots_adjust(wspace=0.01)
sns.boxplot(x='IFUS',y='VotesUS',data=movies,ax=ax1)
plt.title('NON-USA Votes for USA & NON-USA movies')
sns.boxplot(x='IFUS',y='VotesnUS',data=movies,ax=ax2)
plt.tight_layout()
plt.show()
```

**`Inferences:`** Here are the two inferences/observations below:
- Inference 1: Non USA movies are rated less in from people from and not from USA.
- Inference 2: The median rating is same from the people of USA that is value lies between 7.8-8 in both the plots from USA and not-USA movies.
